```python
import numpy as np
import matplotlib.pyplot as plt

# Definir las funciones
def sin_x(x):
    return np.sin(x)

def cos_x(x):
    return np.cos(x)

# Generar un conjunto de valores de x
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calcular los valores de sin(x) y cos(x) para cada x
y_sin = sin_x(x)
y_cos = cos_x(x)

# Crear la gr√°fica
plt.figure(figsize=(8, 6))

# Graficar cada funci√≥n
plt.plot(x, y_sin, label=r'$\sin(x)$')
plt.plot(x, y_cos, label=r'$\cos(x)$')

# Agregar t√≠tulos y etiquetas
plt.title('Gr√°ficas de $\sin(x)$ y $\cos(x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# Mostrar la leyenda
plt.legend()

# Mostrar la gr√°fica
plt.show()
```
